## Dealing with Missing Values in Python

- Data value is stored for a variable (feature) in an observation
- Nan, "?", 0 or a blank space are examples of missing data

## How to deal with missing data?

### Dropping data

This first approach we just remove the entries with some data. However, it only works with Nan values

To this example let's assign a new variable called `df1` and replace the "?" values with Nan values

```python
import numpy as np
df1 = df.replace('?', np.NaN)
```

Then we can drop the data from `df1`

To drop data, we need this:
- The column name (passed to a list)
- axis $\to$ 0 means row and 1 means columns
- inplace $\to$ modify the data frame

```python
df1.dropna(subset=['price'], axis=0)
```

Notice there are less 4 columns than before

### How to replace missing values in Python

As seen before, we can use the `replace` method

Let's replace the nan data with the mean $\downarrow$

```python
df1.dtypes
```

```python
df1['normalized-losses'] = df1['normalized-losses'].astype('float')
```

```python
mean = df1['normalized-losses'].mean()
df1['normalized-losses'].replace(np.NaN, mean)
```

```python
print('Mean of the normalized losses columns: ', mean)
```

## Data Normalization

- The range of the data make the feature impact more or less than others features in, for example, our machine learning models

See the example bellow

```python
df[['length', 'width', 'height']]
```

Notice the range is very different, so making some analysis the `length` feature would impact more. That's why we need to normalize data

Also, it's hard to compare data, see this example:

```python
example1 = pd.DataFrame({
    'age': [20, 30, 40],
    'income': [100000, 20000, 500000]
})
example1
```

Data is not normalized:
- "age" and "income" are in different range
- hard to compare
- "income" will influence the result more

There are several ways to normalize data. Let's see different approaches and techniques

### 1. Simple Feature scaling

$$
\large x_{new}=\dfrac{x_{old}}{x_{max}}
$$

### 2. Min-Max

$$
\large x_{new}=\dfrac{x_{old} - x_{min}}{x_{max} - x_{min}}
$$

### 3. Z-Score


$$
\large x_{new}=\dfrac{x_{old} - \mu}{\sigma}
$$

```python
# Applying simple feature scaling to length
df['length'] / df['length'].max()
```

```python
# Applying Min-Max
(df['length'] - df['length'].min()) / (df['length'].max() - df['length'].min())
```

```python
# Applying Z-Score
(df['length'] - df['length'].mean()) / df['length'].std()
```

## Binning in Python

```python
df.dropna(subset=['price'], axis=0, inplace=True)
```

```python
df
```

Binning data is so simple using python

```python
bins = np.linspace(min(df['price']), max(df['price']), 4)
group_names = ['Low', 'Medium', 'High']
df['price-binned'] = pd.cut(df['price'], bins, labels=group_names, include_lowest=True)
```
